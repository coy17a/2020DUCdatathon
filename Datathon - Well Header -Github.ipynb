```python
import pandas as pd 
import numpy as np

datecolumns = ['LicenceDate', 'ConfidentialReleaseDate','AbandonDate', 'SurfAbandonDate', 'SpudDate', 'FinalDrillDate', 'RigReleaseDate','StatusDate','CompletionDate']   
WellHeader = pd.read_csv(r"C:\Users\ijeke\Documents\Datathon\WellHeader_subset.csv", parse_dates=datecolumns)

WellHeader.info()
```

```python
## create two separate lists of columns to be dropped
dropcodes_1= ["SurfaceOwner", "OSArea", "OSDeposit", "Confidential", "Country", "RegulatoryAgency", "Municipality",  "Agent", "Surf_TownshipRange", "Surf_QuarterUnit", "Surf_Unit", "Surf_Block", "Surf_NTSMapSheet", "Surf_Series", "Surf_Area", "Surf_Sheet", "BH_TownshipRange", "BH_QuarterUnit", "BH_Unit", "BH_Block", "BH_NTSMapSheet", "BH_Series", "BH_Sheet", "BH_Township", "BH_Meridian", "BH_Range", "BH_Section", "BH_LSD", "BH_Area"]

dropcodes_2 = ["CurrentOperatorParent", "CurrentOperatorParentName", "LicenseeParentCompany", "LicenseeParentCompanyName", "WellTypeStandardised", "SurveySystem", "Surf_Township", "Surf_Meridian", "Surf_Range", "Surf_Section", "Surf_LSD"]

dropcodes_3 = ['CompletionDate','_Completed`Flag', '_Perf`Flag', 'Match', 'UnitID', 'UnitFlag', 'UnitName']

##remove the columns based on lists above, and create new dataframe
Well_header_data = WellHeader.drop(columns=(dropcodes_1 + dropcodes_2 + dropcodes_3), axis=1, inplace=False)

Well_header_data.info()
```

```python
##check if EPAssetsId is all unique, by counting number of unique values
uniqueEPAssetsId = Well_header_data['EPAssetsId'].nunique()

print('Number of Unique EPAssetsId: ' + str(uniqueEPAssetsId) )
```

```python
##check if Unamed column 0 is all unique, by counting number of unique values
unique_UnamedColumn0 = Well_header_data.iloc[:, 0].nunique()

print('Number of Unique UnamedColumn0: ' + str(unique_UnamedColumn0) )
```
